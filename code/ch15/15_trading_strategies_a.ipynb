## Simple Moving Averages

### Data Import

```python
import numpy as np
import pandas as pd
import datetime as dt
from pylab import mpl, plt
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
raw = pd.read_csv('http://hilpisch.com/tr_eikon_eod_data.csv',
                  index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
symbol = 'AAPL.O'
```

```python
data = (
    pd.DataFrame(raw[symbol])
    .dropna()
)
```

### Trading Strategy

```python
SMA1 = 42  
SMA2 = 252  
```

```python
data['SMA1'] = data[symbol].rolling(SMA1).mean()  
data['SMA2'] = data[symbol].rolling(SMA2).mean()  
```

```python
data.plot(figsize=(10, 6));
# plt.savefig('../../images/ch15/strat_plot_01.png')
```

```python
data.dropna(inplace=True)
```

```python
data['Position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)  
```

```python
data.tail()
```

```python
ax = data.plot(secondary_y='Position', figsize=(10, 6))
ax.get_legend().set_bbox_to_anchor((0.25, 0.85));
# plt.savefig('../../images/ch15/strat_plot_02.png')
```

### Vectorized Backtesting

```python
data['Returns'] = np.log(data[symbol] / data[symbol].shift(1))  
```

```python
data['Strategy'] = data['Position'].shift(1) * data['Returns']  
```

```python
data.round(4).head()
```

```python
data.dropna(inplace=True)
```

```python
np.exp(data[['Returns', 'Strategy']].sum())  
```

```python
data[['Returns', 'Strategy']].std() * 252 ** 0.5  
```

```python
data[['Returns', 'Strategy']].std()
```

```python
ax = data[['Returns', 'Strategy']].cumsum(
        ).apply(np.exp).plot(figsize=(10, 6))
data['Position'].plot(ax=ax, secondary_y='Position', style='--')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85));
# plt.savefig('../../images/ch15/strat_plot_03.png');
```

### Optimization

```python
from itertools import product
```

```python
# パラメータの取りうる範囲を定める
sma1 = range(20, 61, 4)
sma2 = range(180, 281, 10)  
```

```python
results = pd.DataFrame()
for SMA1, SMA2 in product(sma1, sma2): # SMA1 と SMA2 の全ての組合せに対してループする
    data = pd.DataFrame(raw[symbol]) # DataFrameに変換？
    data.dropna(inplace=True)  # NaN を削除
    data['Returns'] = np.log(data[symbol] / data[symbol].shift(1)) # log return 
    data['SMA1'] = data[symbol].rolling(SMA1).mean()
    data['SMA2'] = data[symbol].rolling(SMA2).mean()
    data.dropna(inplace=True)
    data['Position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
    data['Strategy'] = data['Position'].shift(1) * data['Returns']
    data.dropna(inplace=True)
    perf = np.exp(data[['Returns', 'Strategy']].sum())
    results = results.append(pd.DataFrame(
                {'SMA1': SMA1, 'SMA2': SMA2,
                 'MARKET': perf['Returns'],
                 'STRATEGY': perf['Strategy'],
                 'OUT': perf['Strategy'] - perf['Returns']},
                 index=[0]), ignore_index=True)  
```

```python
results.info()
```

```python
# 'OUT' が大きいほどパフォーマンスがよい
# 'OUT'の降順に並べたら先頭に一番よいパラメータがくる
results.sort_values('OUT', ascending=False).head(7)
```

## Random Walk Hypothesis

```python
# S&P500指数を扱う
symbol = '.SPX'
```

```python
# DataFrame に変換
data = pd.DataFrame(raw[symbol])
```

```python
lags = 5
cols = []
for lag in range(1, lags + 1):
    col = 'lag_{}'.format(lag)  
    data[col] = data[symbol].shift(lag)  
    cols.append(col)  
```

```python
data.head(7)
```

```python
data.dropna(inplace=True)
```

```python
# 最小二乗法で最小二乗解を求める
reg = np.linalg.lstsq(data[cols], data[symbol], rcond=-1)[0]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(cols, reg);
# plt.savefig('../../images/ch15/strat_plot_04.png');
```

```python
# 最小二乗解を遅延データ系列に適用
data['Prediction'] = np.dot(data[cols], reg)
```

```python
data[[symbol, 'Prediction']].iloc[-75:].plot(figsize=(10, 6));
# plt.savefig('../../images/ch15/strat_plot_05.png');
```
